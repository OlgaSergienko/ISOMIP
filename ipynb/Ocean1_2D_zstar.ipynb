```python
from matplotlib import pyplot as plt
%matplotlib inline
import sys
sys.path.append('../plotting/')
from plotting import *
import netCDF4
import numpy as np
```

```python
path = '/glade/work/gmarques/Projects/ISOMIP/setup/Ocean1_2D_zstart/' 
```

```python
prog_file='prog_0001.nc'
ic_file='ISOMIP_IC.nc'
exps = ['A','B','C','D']
```

## Initial conditions

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(16,10), sharey=True, sharex=True)
ax = ax.flatten()
for i in range(len(exps)):
  ic = netCDF4.Dataset(path+exps[i]+'/'+ic_file)
  xq = ic.variables['lonq'][:] # This is the coordinate of the cell corners (u-points in 1D)
  plot_section(ic, 0, xq, variable='Salt',eta='eta',ax=ax[i]); ax[i].set_title(str(exps[i]), fontsize=18);
```

### Eastward Velocity

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(16,10), sharey=True, sharex=True)
ax = ax.flatten()
for i in range(len(exps)):
  prog = netCDF4.Dataset(path+exps[i]+'/'+prog_file)
  print('Time = {} days'.format(str(prog.variables['time'][-1])))
  xq = prog.variables['xq'][:] # This is the coordinate of the cell corners (u-points in 1D)
  plot_section(prog, -1, xq, variable='u', ax=ax[i],clim=(-1e-4,1e-4), cmap=plt.cm.bwr) 
  ax[i].set_title(str(exps[i]), fontsize=18);
```

### Northward Velocity

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(16,10), sharey=True, sharex=True)
ax = ax.flatten()
for i in range(len(exps)):
  print('Time = {} days'.format(str(prog.variables['time'][0])))
  prog = netCDF4.Dataset(path+exps[i]+'/'+prog_file)
  xq = prog.variables['xq'][:] # This is the coordinate of the cell corners (u-points in 1D)
  plot_section(prog, 0, xq, variable='v', ax=ax[i],clim=(-1e-2,1e-2), cmap=plt.cm.bwr) 
  ax[i].set_title(str(exps[i]), fontsize=18);
```

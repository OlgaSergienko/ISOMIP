# Cold forcing
This test should not generate any motion since T/S values identical to the IC are applied in the sponge layer.

```python
from matplotlib import pyplot as plt
%matplotlib inline
import sys
sys.path.append('../plotting/')
from plotting import *
import netCDF4
import numpy as np
```

```python
# plot IC
path='/lustre/f1/unswept/Gustavo.Marques/MOM6-examples/ocean_only/ISOMIP/TESTS/2D/SPONGES/COLD/'
icfile = 'ISOMIP_IC.nc'
layer_file = netCDF4.Dataset(path+'layer/'+icfile)
rho_file = netCDF4.Dataset(path+'rho/'+icfile)
sigma_file = netCDF4.Dataset(path+'sigma/'+icfile)
z_file = netCDF4.Dataset(path+'z/'+icfile)
# Read the horizontal coordinate which is the same for all configurations 
xq = layer_file.variables['latq'][:] # This is the coordinate of the cell corners (u-points in 1D)
xq = np.concatenate(([0],xq)) # Inserts left most edge of domain in to coordinate
record = 0
plt.figure(figsize=(18,9))
plt.subplot(2,2,1); plot_section(layer_file, record, xq, i=2,variable='Salt', eta='eta'); plt.title('Layer', fontsize=18);
plt.subplot(2,2,2); plot_section(rho_file, record, xq, i=2,variable='Salt', eta='eta'); plt.title(r'$\rho$', fontsize=18);
plt.subplot(2,2,3); plot_section(sigma_file, record, xq, i=2,variable='Salt', eta='eta'); plt.title(r'$\sigma$', fontsize=18);
plt.subplot(2,2,4); plot_section(z_file, record, xq, i=2,variable='Salt', eta='eta'); plt.title(r'z*', fontsize=18);
plt.tight_layout()
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
import sys
sys.path.append('../plotting/')
from plotting import *
import netCDF4
import numpy as np
```

```python
# Open the output from the four experiments
path='/lustre/f1/unswept/Gustavo.Marques/MOM6-examples/ocean_only/ISOMIP/TESTS/2D/YES_IS/'
layer_file = netCDF4.Dataset(path+'layer/prog__0001_002.nc')
rho_file = netCDF4.Dataset(path+'rho/prog__0001_002.nc')
sigma_file = netCDF4.Dataset(path+'sigma/prog__0001_002.nc')
z_file = netCDF4.Dataset(path+'z/prog__0001_002.nc')
```

```python
# Read the horizontal coordinate which is the same for all configurations 
xq = layer_file.variables['xq'][:] # This is the coordinate of the cell corners (u-points in 1D)
xq = np.concatenate(([0],xq)) # Inserts left most edge of domain in to coordinate
```

```python
record = -1 # Last record
print 'Time is:',layer_file.variables['Time'][record], 'days'
plt.figure(figsize=(18,9))
plt.subplot(2,2,1); plot_section(layer_file, record, xq); plt.title('Layer', fontsize=18);
plt.subplot(2,2,2); plot_section(rho_file, record, xq); plt.title(r'$\rho$', fontsize=18);
plt.subplot(2,2,3); plot_section(sigma_file, record, xq); plt.title(r'$\sigma$', fontsize=18);
plt.subplot(2,2,4); plot_section(z_file, record, xq); plt.title(r'$z^*$', fontsize=18);
plt.tight_layout()
```

```python
variable = 'u'
minmax=(-1e-2,1e-2)
plt.figure(figsize=(18,9))
plt.subplot(2,2,1); plot_section(layer_file, record, xq, variable=variable, clim=(-1e-8,1e-8)); plt.title('Layer', fontsize=18);
plt.subplot(2,2,2); plot_section(rho_file, record, xq, variable=variable, clim=minmax); plt.title(r'$\rho$', fontsize=18);
plt.subplot(2,2,3); plot_section(sigma_file, record, xq, variable=variable, clim=minmax); plt.title(r'$\sigma$', fontsize=18);
plt.subplot(2,2,4); plot_section(z_file, record, xq, variable=variable, clim=(-1e-5,1e-5)); plt.title(r'$z^*$', fontsize=18);
plt.tight_layout()
```
